```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dir = Path('/standard/BDShackathon/data/leukemia/clinical')

med_raw = pd.read_csv(dir/'20250317_UVA_Medications_V4.csv')
display(med_raw)

transplant_raw = pd.read_csv(dir/'20250317_UVA_StemCellTransplant_V4.csv')
display(transplant_raw)
```

```python
treatments = (
    pd.get_dummies(med_raw.set_index('AvatarKey')['Medication'])
    .reset_index()
    .groupby('AvatarKey')
    .agg(any)
    .drop(columns='Unknown/Not Applicable')
)
treatments['StemCellTransplant'] = transplant_raw.set_index('AvatarKey')['SCTInd'] == 'Yes'
treatments.to_csv('./cleaned_clinical_data_csvs/treatments.csv', sep=',')
treatments
```

```python
treatment_count = treatments.sum(axis=0).sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(12, 4))
ax = sns.barplot(treatment_count, ax=ax)
ax.set(ylabel='# of patients')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

```python
number_of_prescriptions = (
    med_raw[['AvatarKey', 'Medication']]
    .groupby('AvatarKey')
    .agg('count')
)['Medication'].rename('NumberOfMedications')
number_of_prescriptions.to_csv('./cleaned_clinical_data_csvs/number_of_prescriptions.csv', sep=',')

ax = sns.histplot(number_of_prescriptions, discrete=True)
ax.set(ylabel='# of patients')
```

```python
med_transplant = pd.merge(treatments.drop(columns='StemCellTransplant').any(axis=1).rename('Medication'), treatments['StemCellTransplant'], left_index=True, right_index=True)
med_transplant.value_counts()
```
